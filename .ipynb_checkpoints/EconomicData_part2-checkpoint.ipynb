```python
import pandas as pd
import numpy as np
```

```python
file_name = "./data/Economic Data.xlsx"
```

```python
xl = pd.ExcelFile(file_name)
for sheet in xl.sheet_names:
    print(sheet)
```

```python
df = xl.parse("Depart vs Arrival by Port Px")
```

```python
df = df.iloc[1:,:5]
```

```python
df.columns = ["Date", "PortID", "PxType", "Arrivals", "Departures"]
```

```python
portIDMap = {1: "Airport", 2: "Express Rail Link West Kowloon", 3: "Hung Hom", 4: "Lo Wu", 5: "Lok Ma Chau Spur Line", 6: "Hong Kong-Zhuhai-Macao Bridge", 7: "Lok Ma Chau", 8: "Man Kam To", 9: "Sha Tau Kok", 10: "Shenzhen Bay", 11: "China Ferry Terminal", 12: "Harbour Control", 13: "Kai Tak Cruise Terminal", 14: "Macau Ferry Terminal", 15: "Tuen Mun Ferry Terminal", 16: "Heung Yuen Wai"}
```

```python
pxTypeMap = {1: "Hong Kong Residents", 2: "Mainland Visitors", 3: "Other Visitors"}
```

```python
portIDMap[1]
```

```python
for idx, row in df.iterrows():
    port_id = row.PortID
    px_type = row.PxType
    df.loc[idx, "PortIDName"] = portIDMap[port_id]
    df.loc[idx, "PxTypeName"] = pxTypeMap[px_type]
```

```python
df
```

```python
df["Date"] = df["Date"].map(str)
```

```python
df["Year"] = df["Date"].str[:4].map(int)
```

```python
df["Month"] = df["Date"].str[5:7].map(int)
```

```python
df["Day"] = df["Date"].str[8:10].map(int)
```
