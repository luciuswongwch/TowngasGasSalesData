```python
import pandas as pd
```

```python
file_name = "./data/Gas Sales by Trade Code.xlsx"
```

```python
xl = pd.ExcelFile(file_name)
for sheet in xl.sheet_names:
    df = xl.parse(sheet)
    df.head()
```

```python
df.head()
```

```python
#df = df.fillna(method="ffill")
```

```python
last_year = "2015"
for idx, row in enumerate(df.values):
    if (str(row[0]) != "nan"):
        last_year = str(row[0])
    df.iloc[idx][0] = last_year
```

```python
df = df.fillna(method="ffill", axis=1)
```

```python
df = df.fillna(0)
```

```python
df
```

```python
import calendar
```

```python
def month_number(month: str):
    for c_index, c_month in enumerate(calendar.month_abbr):
        if (month[:3] == c_month):
            return c_index
```

```python
columns = ["Year", "Month", "MonthNumber", "TradeGroupType", "TradeGroup", "TradeGroupCode", "GasSales", "Description"]
```

```python
#df.columns = df.iloc(1)
```

```python
#df.drop(1)
```

```python
result_df = pd.DataFrame()
```

```python
for idx, trade_group in enumerate(df.iloc[1][2:]):
    for row in df[2:].values:
        trade_group_code = trade_group.split("-")[0].strip()
        if (trade_group_code[0] not in ["C", "I"]):
            trade_group_code = "I" + trade_group_code
        new_row = [row[0], row[1], month_number(row[1]), df.iloc[0][idx+2], trade_group, trade_group_code, row[idx+2], "Gas Sales in TJ"]
        result_df = pd.concat([result_df, pd.DataFrame([new_row])], ignore_index=True)
```

```python
result_df.columns = columns
```
