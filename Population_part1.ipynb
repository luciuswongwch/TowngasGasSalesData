```python
df = xl.parse("E101e")
```

```python
df = df.iloc[1:11,:7]
```

```python
columns = ["Year", "TypeOfHousing", "NumberOfPersons", "PercentageOfPopulation", "Description"]
```

```python
df.drop(2, inplace=True)
```

```python
df = df.fillna(method="ffill", axis=1)
```

```python
df
```

```python
result_df = pd.DataFrame()
```

```python
for idx, year in enumerate(df.iloc[0, 1:].drop_duplicates()):
    for inner_idx, housingType in enumerate(df.iloc[3:,0]):
        new_row = [year, housingType, df.iloc[inner_idx + 3, idx * 2 + 1], df.iloc[inner_idx + 3, idx * 2 + 2], "Population by housing type and year"]
        result_df = pd.concat([result_df, pd.DataFrame([new_row])], ignore_index=True)
```

```python
result_df.columns = columns
```
