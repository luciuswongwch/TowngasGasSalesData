```python
import calendar
```

```python
def month_number(month: str):
    for c_index, c_month in enumerate(calendar.month_abbr):
        if (month[:3] == c_month):
            return c_index
```

```python
df["MonthNumber"] = df["Month"].map(month_number)
```

```python
df2 = xl.parse("Gas Sales  by Trade")
```

```python
df2 = df2.fillna(method="ffill")
```

```python
df2 = df2.fillna(method="ffill", axis=1)
```

```python
df2.iloc[0, -1] = "Commercial and Industrial"
```

```python
df2
```

```python
def month_number(month: str):
    for c_index, c_month in enumerate(calendar.month_abbr):
        if (month[:3] == c_month):
            return c_index
```

```python
result_df = pd.DataFrame()
```

```python
for idx, tradeGroup in enumerate(df2.iloc[0,2:]):
    for inner_idx, row in enumerate(df2.iterrows()):
        if (inner_idx not in [0, 1]):
            new_row = [df2.iloc[inner_idx, 0], df2.iloc[inner_idx, 1], month_number(df2.iloc[inner_idx, 1]), df2.iloc[1, 2 + idx], tradeGroup, df2.iloc[inner_idx, 2+idx], "Gas Sales in TJ"]
            result_df = pd.concat([result_df, pd.DataFrame([new_row])], ignore_index=True)
```

```python
result_df.columns = ["Year", "Month", "MonthNumber", "TradeGroup", "TradeGroupType", "MonthlyGasSales", "Description"]
```

```python
result_df
```

```python
cursor = cnxn.cursor()
# Insert Dataframe into SQL Server:
for index, row in result_df.iterrows():
    cursor.execute("INSERT INTO EconomicData.GasSalesByTrade (Year, Month, MonthNumber, TradeGroup, TradeGroupType, MonthlyGasSales, Description) values(?,?,?,?,?,?,?)", row.Year, row.Month, row.MonthNumber, row.TradeGroup, row.TradeGroupType, toEmptyOrFloat(row.MonthlyGasSales), row.Description) 
cnxn.commit()
cursor.close()
```
